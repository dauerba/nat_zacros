### nat_zacros SimulationRun Example

This notebook demonstrates how to test the `SimulationRun` class from the `nat_zacros` package in an interactive, step-by-step manner.

### 1. Set up the Python Environment

Import essential libraries and import nat-zacros

```python
import json
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
from   pathlib import Path
import pickle
import platform
import time
import sys
# import pandas as pd

try:
    import nat_zacros as nz
    print(f'nat_zacros successfully imported from: {nz.__file__}')
    print(f'nat_zacros version: {nz.__version__}\n')

except ImportError:  
    raise ImportError(
    'nat_zacros must be installed in the conda or virtual environment in use '
    'to run this example successfully.\n'
    'see nat-zacros packageINSTALLATION.md for instructions')

print(f'using conda virtual environment: {Path(sys.executable).parent}')
# print(f"python version: {sys.version}")
# print(f"Python executable: {sys.executable}")
```

### 2. Use nz.SimulationRun and create an example run

```python
# Use an example directory with zacros calculations 
#       To do: document the directory structure we are using for the data

example_dir  = Path.home()/'Dropbox'/'Surface_Reaction_Kinetics'/'O_Pt111'/'zacros_calculations'
example_data = Path('fn_3leed') / 'jobs' / '1'

try:
    run1 = nz.SimulationRun(example_dir / example_data)
    print(f"SimulationRun created for: {example_data}")
except Exception as e:
    print(f"Error: {e}")

print("Run name:              ", run1.run_dir.name)
print("Number of trajectories:", len(run1.traj_dirs))
print("Lattice size:          ", run1.lattice.size)
```
